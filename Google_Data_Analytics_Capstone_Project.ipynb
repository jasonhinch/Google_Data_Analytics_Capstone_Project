# 3. Process

### Transform

Transform the data so you can work with it effectively

```r
# Load necessary packages

library("tidyverse")
library("ggplot2")
library("lubridate")
```

```r
# Create variables for data

tripdata_2020_04 <- read.csv("../input/cyclistic/202004-divvy-tripdata/202004-divvy-tripdata.csv")
tripdata_2020_05 <- read.csv("../input/cyclistic/202005-divvy-tripdata/202005-divvy-tripdata.csv")
tripdata_2020_06 <- read.csv("../input/cyclistic/202006-divvy-tripdata/202006-divvy-tripdata.csv")
tripdata_2020_07 <- read.csv("../input/cyclistic/202007-divvy-tripdata/202007-divvy-tripdata.csv")
tripdata_2020_08 <- read.csv("../input/cyclistic/202008-divvy-tripdata/202008-divvy-tripdata.csv")
tripdata_2020_09 <- read.csv("../input/cyclistic/202009-divvy-tripdata/202009-divvy-tripdata.csv")
tripdata_2020_10 <- read.csv("../input/cyclistic/202010-divvy-tripdata/202010-divvy-tripdata.csv")
tripdata_2020_11 <- read.csv("../input/cyclistic/202011-divvy-tripdata/202011-divvy-tripdata.csv")
tripdata_2020_12 <- read.csv("../input/cyclistic/202012-divvy-tripdata/202012-divvy-tripdata.csv")
tripdata_2021_01 <- read.csv("../input/cyclistic/202101-divvy-tripdata/202101-divvy-tripdata.csv")
tripdata_2021_02 <- read.csv("../input/cyclistic/202102-divvy-tripdata/202102-divvy-tripdata.csv")
tripdata_2021_03 <- read.csv("../input/cyclistic/202103-divvy-tripdata/202103-divvy-tripdata.csv")
```

```r
# Join the data from tripdata_2020_04 to tripdata_2020_11

tripdata_withInt <- bind_rows(tripdata_2020_04,tripdata_2020_05,tripdata_2020_06,tripdata_2020_07,tripdata_2020_08,tripdata_2020_09,tripdata_2020_10,tripdata_2020_11)

# Join the data from tripdata_2020_12 to tripdata_2021_03

tripdata_withChar <- bind_rows(tripdata_2020_12,tripdata_2021_01,tripdata_2021_02,tripdata_2021_03)

# Change the type of two tripdata_withInt columns (start_station_id,end_station_id) to match the tripdata_withChar column type

tripdata_withInt <-  mutate(tripdata_withInt, start_station_id = as.character(start_station_id), end_station_id = as.character(end_station_id)) 

# Join all the data:

all_tripdata <- bind_rows(tripdata_withInt, tripdata_withChar)
```

```r
glimpse(all_tripdata)
```

```r
summary(all_tripdata)
```

## Cleaning:

Clean the data and document the cleaning process.

```r
# Remove all incomplete rows with missing/NA values

all_tripdata_no_na <- drop_na(all_tripdata)
```

```r
# Remove duplicates

all_tripdata_clean <- all_tripdata_no_na[!duplicated(all_tripdata_no_na$ride_id), ]
print(paste("Removed", nrow(all_tripdata_no_na) - nrow(all_tripdata_clean), "duplicated rows"))
```

```r
# Bike maintenance by Divvy

all_tripdata_clean <- all_tripdata_clean[!(all_tripdata_clean$start_station_name == "HQ QR"),]
```

```r
# Parse datetime columns

all_tripdata_clean$started_at <- as.POSIXct(all_tripdata_clean$started_at, "%Y-%m-%d %H:%M:%S")
all_tripdata_clean$ended_at <- as.POSIXct(all_tripdata_clean$ended_at, "%Y-%m-%d %H:%M:%S")
```

### Create useful columns for later:

Create columns for future calculations and visualisations.

```r
# Ride time in minutes

all_tripdata_clean <- all_tripdata_clean %>%
    mutate(ride_time_m = as.numeric(all_tripdata_clean$ended_at - all_tripdata_clean$started_at) / 60)

summary(all_tripdata_clean$ride_time_m)
```

```r
# Separate the year and the month into a new column

all_tripdata_clean <- all_tripdata_clean %>%
    mutate(year_month = paste(strftime(all_tripdata_clean$started_at, "%Y"),
                              "-",
                              strftime(all_tripdata_clean$started_at, "%m"),
                              paste("(",strftime(all_tripdata_clean$started_at, "%b"), ")", sep="")))

unique(all_tripdata_clean$year_month)
```

```r
# Weekday to determine patterns of travels in the week

all_tripdata_clean <- all_tripdata_clean %>%
    mutate(weekday = paste(strftime(all_tripdata_clean$ended_at, "%u"), "-", strftime(all_tripdata_clean$ended_at, "%a")))

unique(all_tripdata_clean$weekday)
```

```r
# Starting hour of each ride

all_tripdata_clean <- all_tripdata_clean %>%
    mutate(start_hour = strftime(all_tripdata_clean$ended_at, "%H"))

unique(all_tripdata_clean$start_hour)
```

```r
# This function help to resize the plots

fig <- function(width, heigth){options(repr.plot.width = width, repr.plot.height = heigth)}

fig(16,8)
```

# 4. Analyze

### Casuals vs Members:

```r
all_tripdata_clean %>% 
    group_by(member_casual) %>% 
    summarise(count = length(ride_id),
              '%' = (length(ride_id) / nrow(all_tripdata_clean)) * 100)
```

```r
ggplot(all_tripdata_clean, aes(member_casual, fill=member_casual)) +
    geom_bar() +
    labs(x="Casual vs Member", title="Data viz 01 - Casual vs Member distribution") +
    coord_flip()
```

### Distribution by month

```r
all_tripdata_clean %>%
    group_by(year_month) %>%
    summarise(count = length(ride_id),
              '%' = (length(ride_id) / nrow(all_tripdata_clean)) * 100,
              'members_p' = (sum(member_casual == "member") / length(ride_id)) * 100,
              'casual_p' = (sum(member_casual == "casual") / length(ride_id)) * 100,
              'Member x Casual Perc Difer' = members_p - casual_p)
```

```r
all_tripdata_clean %>%
  ggplot(aes(year_month, fill=member_casual)) +
    geom_bar() +
    labs(x="Month", title="Data viz 02 - Distribution by month")
```

### Distribution by weekday:

```r
all_tripdata_clean %>%
    group_by(weekday) %>% 
    summarise(count = length(ride_id),
              '%' = (length(ride_id) / nrow(all_tripdata_clean)) * 100,
              'members_p' = (sum(member_casual == "member") / length(ride_id)) * 100,
              'casual_p' = (sum(member_casual == "casual") / length(ride_id)) * 100,
              'Member x Casual Perc Difer' = members_p - casual_p)
```

```r
ggplot(all_tripdata_clean, aes(fct_rev(weekday), fill=member_casual)) +
    geom_bar() +
    labs(x="Weekdady", title="Data viz 03 - Distribution by weekday") +
    coord_flip()
```

### Distribution by hour:

```r
all_tripdata_clean %>%
    group_by(start_hour) %>% 
    summarise(count = length(ride_id),
          '%' = (length(ride_id) / nrow(all_tripdata_clean)) * 100,
          'members_p' = (sum(member_casual == "member") / length(ride_id)) * 100,
          'casual_p' = (sum(member_casual == "casual") / length(ride_id)) * 100,
          'member_casual_perc_difer' = members_p - casual_p)
```

```r
all_tripdata_clean %>%
    ggplot(aes(start_hour, fill=member_casual)) +
    labs(x="Hour of the day", title="Data viz 04 - Distribution by hour of the day") +
    geom_bar()
```

### Facet_wrap by weekday:

```r
all_tripdata_clean %>%
    ggplot(aes(start_hour, fill=member_casual)) +
    geom_bar() +
    labs(x="Hour of the day", title="Data viz 05 - Distribution by hour of the day divided by weekday") +
    facet_wrap(~ weekday)
```

### Facet_wrap by weekend vs weekday:

```r
all_tripdata_clean %>%
    mutate(type_of_weekday = ifelse(weekday == '6 - Sat' | weekday == '7 - Sun',
                                   'Weekend',
                                   'Weekday')) %>%
    ggplot(aes(start_hour, fill=member_casual)) +
    labs(x="Hour of the day", title="Data viz 06 - Distribution by hour of the day on weekend vs weekday") +
    geom_bar() +
    facet_wrap(~ type_of_weekday)
```

### Rideable type:

```r
all_tripdata_clean %>%
    group_by(rideable_type) %>% 
    summarise(count = length(ride_id),
          '%' = (length(ride_id) / nrow(all_tripdata_clean)) * 100,
          'members_p' = (sum(member_casual == "member") / length(ride_id)) * 100,
          'casual_p' = (sum(member_casual == "casual") / length(ride_id)) * 100,
          'member_casual_perc_difer' = members_p - casual_p)
```

```r
ggplot(all_tripdata_clean, aes(rideable_type, fill=member_casual)) +
    labs(x="Rideable type", title="Data viz 07 - Distribution of types of bikes") +
    geom_bar() +
    coord_flip()
```

### Ride time summary:

```r
summary(all_tripdata_clean$ride_time_m)
```

```r
# Examine outliers

ventiles = quantile(all_tripdata_clean$ride_time_m, seq(0, 1, by=0.05))
ventiles
```

```r
# Remove outliers 

tripdata_without_outliers <- all_tripdata_clean %>% 
    filter(ride_time_m > as.numeric(ventiles['5%'])) %>%
    filter(ride_time_m < as.numeric(ventiles['95%']))

print(paste("Removed", nrow(all_tripdata_clean) - nrow(tripdata_without_outliers), "rows as outliners" ))
```

```r
# Examine new ride time summary without outliers

tripdata_without_outliers %>% 
    group_by(member_casual) %>% 
    summarise(mean = mean(ride_time_m),
              'first_quarter' = as.numeric(quantile(ride_time_m, .25)),
              'median' = median(ride_time_m),
              'third_quarter' = as.numeric(quantile(ride_time_m, .75)),
              'IR' = third_quarter - first_quarter)
```

### Ride time boxplots:

```r
ggplot(tripdata_without_outliers, aes(x=member_casual, y=ride_time_m, fill=member_casual)) +
    labs(x="Casual vs Member", y="Riding time", title="Data viz 08 - Distribution of Riding time for Casual vs Member") +
    geom_boxplot() +
    coord_flip()
```

### Ride time facet_wrap:

```r
ggplot(tripdata_without_outliers, aes(x=fct_rev(weekday), y=ride_time_m, fill=member_casual)) +
    geom_boxplot() +
    facet_wrap(~ member_casual) +
    labs(x="Weekday", y="Riding time", title="Data viz 09 - Distribution of Riding time for day of the week")
```

### Rideable type facet_wrap:

```r
ggplot(tripdata_without_outliers, aes(x=rideable_type, y=ride_time_m, fill=member_casual)) +
    geom_boxplot() +
    facet_wrap(~ member_casual) +
    labs(x="Rideable type", y="Riding time", title="Data viz 10 - Distribution of Riding time for rideable type")
```

## Recommendations 

### To drive casual members to become members:
1. Create weekly and monthly passes as a middle ground for casuals to become members.
2. Introduce surge pricing for casual riders during peak hours (summer/weekends) and at the busiest stations ie: Central and North Chicago.

5. Introduce a discounted off peak membership which allows for use only during off peak hours.
